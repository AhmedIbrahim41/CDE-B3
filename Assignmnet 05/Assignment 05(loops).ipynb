## Problem 15


```python
base_number = float(input("enter the base number"))
expo_number = int(input("enter the exponent number"))
ans = 1
if expo_number > 0:
    for i in range(1,expo_number +1 ):
        ans *= base_number
else:
    for i in range(1,-expo_number +1 ):
        ans *= base_number
    ans = 1 /ans

print(ans)
```
