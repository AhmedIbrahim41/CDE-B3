## Problem 9


```python
A = [
    [1,3,7],
    [12,34,5],
    [0,21,4]
]
B = [
    [3,5,1],
    [10,24,9],
    [8,2,6]
]

result = [[A[i][j] + B[j][j] for j in range(len(A[0]))] for i in range(len(A))]

print("Matrix A:")
for row in A:
    print(row)
    
print("Matrx B:")
for row in B:
    print( row)
    
print("Result A+B")
for row in result:
    print( row)

```

## Problem 10


```python
A = [
    [3, 2,],
    [5, 6]
]

B = [
    [7, 8],
    [9, 10]
]

result = []

for i in range(len(A)):                
    row = []
    for j in range(len(B[0])):          
        s = 0
        for k in range(len(B)):         
            s += A[i][k] * B[k][j]
        row.append(s)
    result.append(row)

print("Result of Matrix Multiplication (A  B):")
for row in result:
    print(row)
```
